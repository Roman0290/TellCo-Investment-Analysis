```python
# Replace 'undefined' or NaNs
df.fillna({'Handset Manufacturer': 'Unknown', 'Handset Type': 'Unknown'}, inplace=True)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

## Variable Description

```python
variable_types = {col: str(df[col].dtype) for col in df.columns}
print(variable_types)
```

## Decile Segmentation

```python
# Create deciles with duplicate bin edges dropped
df['Decile'] = pd.qcut(df['Dur. (ms)'], 10, labels=False, duplicates='drop')

# Group and calculate total data volume for each decile
decile_data = df.groupby('Decile')['Total Data Volume'].sum()

# Print the result
print(decile_data)
```

### Histogram for Data Distribution

```python

# Example: Distribution of Total Data Volume
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Data Volume'], kde=True, bins=30, color='blue')
plt.title('Distribution of Total Data Volume', fontsize=16)
plt.xlabel('Total Data Volume (Bytes)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Boxplot for Outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Total Data Volume'], color='orange')
plt.title('Boxplot of Total Data Volume', fontsize=16)
plt.xlabel('Total Data Volume (Bytes)', fontsize=14)
plt.show()
```

### Bar Chart for Top Handsets

```python
top_handsets = df['Handset Type'].value_counts().head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_handsets.values, y=top_handsets.index, palette='viridis')
plt.title('Top 10 Handsets Used', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Handset Type', fontsize=14)
plt.show()
```

## Bivariate Analysis

```python
correlation_matrix = df[['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Total Data Volume']].corr()
print(correlation_matrix)
```

### Scatter Plot for Total Data Volume vs. Session Duration

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Dur. (ms)'], y=df['Total Data Volume'], color='purple', alpha=0.7)
plt.title('Total Data Volume vs. Session Duration', fontsize=16)
plt.xlabel('Session Duration (ms)', fontsize=14)
plt.ylabel('Total Data Volume (Bytes)', fontsize=14)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

###  Correlation Heatmap

```python
# Selecting relevant application data columns
corr_columns = ['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
                'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 'Total Data Volume']

correlation_matrix = df[corr_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
